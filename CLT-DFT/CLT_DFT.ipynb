#Normal dağılım

```
from numpy.random import seed,randint
from numpy import mean
from scipy.stats import norm
from matplotlib import pyplot as plt
import numpy as np
seed(1)

x_axis = np.arange(0, 6, 0.001)
plt.plot(x_axis, norm.pdf(x_axis,3))
plt.show()
```

**3 adet zarın atılması ile oluşan ortalamalar grafiği**

```
seed(1)
ortalama = [mean(randint(1, 7, 3)) for _ in range(1000)]

plt.hist(ortalama)
plt.show()

```

**4 adet zarın atılması ile oluşan ortalamalar grafiği**

```
seed(1)
ortalama = [mean(randint(1, 7, 4)) for _ in range(1000)]

plt.hist(ortalama)
plt.show()
```

**5 adet zarın atılması ile oluşan ortalamalar grafiği**

```
seed(1)
ortalama = [mean(randint(1, 7, 5)) for _ in range(1000)]

plt.hist(ortalama)
plt.show()
```

  3,4 ve 5 adet zar sırası ile 1000'ner defa birbirinden bağımsız bir şekilde atılıp, çıkan sonuçların aritmetik ortalamaları hesaplanmıştır. Bu hesaplanana değerler histogramlar yardımı ile görselleştirilmiştir. 

  Bu histogramlar incelendiğinde ,   bir birinden bağımsız ve aynı dağılımın parçası olan bu rassal değişkenlerin aritmetik ortalamalarının , örnekleme sayısı artıkça , dağılımı "normal" dağılıma benzerlik göstermektedir. Bu duruma "Merkezi Limit Teoremi" denilmektedir.

#Ayrik fourier transformunu

**DFT**
![](https://wikimedia.org/api/rest_v1/media/math/render/svg/18b0e4c82f095e3789e51ad8c2c6685306b5662b)



```
import numpy as np 
def dft(x):# x sinyal    
    N = x.size
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    return np.dot(e, x)
```

```
t = np.arange(0,2,0.01)
x = np.sin(2*np.pi*10*t)+np.sin(2*np.pi*40*t)# 10 + 40 hz
import matplotlib.pyplot as plt
plt.plot(t,x)
plt.show()
```

```
dft_x  = dft(x)

#print(dft_x)
#X = [x.real for x in dft_x]
#Y = [x.imag for x in dft_x]


N = dft_x.size
print(N)
f = np.linspace(0, 1 / (0.01), N)
plt.ylabel("Genlik")
plt.xlabel("Frekans ")
#https://stackoverflow.com/questions/25735153/plotting-a-fast-fourier-transform-in-python
plt.bar(f[:N // 2], np.abs(dft_x)[:N // 2] * 1 / N, width=1.5)  # 1 / N is a normalization factor

plt.show()
```
